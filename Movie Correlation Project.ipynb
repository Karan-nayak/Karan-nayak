```python
# First let's import the packages we will use in this project
# You can do this all now or as you need them
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #adjusts the configuration of the plots we will create


```

```python
# Now we need to read in the data
df = pd.read_csv(r"C:\Users\yani\Downloads\movies.csv")
```

```python
df.head()
```

```python
# We need to see if we have any missing data
# Let's loop through the data and see if there is anything missing

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Data Types for our columns

print(df.dtypes)
```

```python
# change data type of columns

df['budget']= df['budget'].astype('int64')
df['gross']= df['gross'].astype('int64')
```

```python
df = df.sort_values(by=['gross'], inplace = False, ascending=False)
```

```python
#drop any duplicates

df.drop_duplicates()
```

```python
#budget high correlation
#company high corelation
```

```python
#scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('budget vs gross Earnings')
plt.xlabel('Gross earnings')
plt.ylabel('budget for film')

plt.show()
```

```python
df.head()
```

```python
# plot budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color": "blue"})
```

```python
# lets start looking at correlation
#this correlation will only be working on column with numerical values

df.corr(method ='pearson')  #pearson - by default, kendall, spearman are diff methods of corrrelation
```

```python
#high correlation between budget and gross

#bright correlation is high and dark represent low correlation

correlation_matrix= df.corr('pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('correlation for numeric feature')
plt.xlabel('movie features')
plt.ylabel('movie features')

plt.show()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype=='object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name]= df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python

correlation_matrix= df_numerized.corr('pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('correlation for numeric feature')
plt.xlabel('movie features')
plt.ylabel('movie features')

plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs)>0.5]

high_corr
```
